```python
import numpy as np
import control
import matplotlib.pyplot as plt

# Define System
num = [1, 1]
den = [1, 4, 7, 6, 3]
system = control.TransferFunction(num, den)
```

```python
# Transient Analysis (Step Response)
t, y = control.step_response(system)
plt.figure(figsize=(10, 6))
plt.plot(t, y, color='#2563eb', linewidth=2)
plt.title("Step Response of 2-DOF Mechanical System", fontsize=14, pad=20)
plt.xlabel("Time (s)")
plt.ylabel("Displacement X2")
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('step_response.png')
```

```python
# Stability Analysis (Pole-Zero Map)
plt.figure(figsize=(8, 8))
control.pzmap(system, plot=True)
plt.title("Pole-Zero Map for Stability Verification", fontsize=14, pad=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('pz_map.png')

print(f"System Stability: {'Stable' if np.all(np.real(control.poles(system)) < 0) else 'Unstable'}")
```
