```python
import sympy as sp
s, S = sp.symbols('s S') 

# Coefficients only (no X,Y)
A_coeff = 2*S**2 + s      # 2S² + s
B_coeff = s + 1           # s + 1

sp.pprint(sp.simplify(B_coeff / A_coeff))
```

```python
from control import tf, poles, zeros

# Direct coefficients (no sympy needed!)
num_coeffs = [1, 1]   # s + 1
den_coeffs = [1, 2]   # s + 2

sys = tf(num_coeffs, den_coeffs)
print(sys)
print(f"Poles: {poles(sys)}")
print(f"Zeros: {zeros(sys)}")
```

```python
import control as ctrl
import matplotlib.pyplot as plt

# NEW TRANSFER FUNCTION for gain=25
G = ctrl.tf([25, 25], [1, 2])  # 25(s+1)/(s+2) → Steady-state=25

# Your exact code:
t, y = ctrl.step_response(G)
plt.figure()   
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid()
plt.show()
```
